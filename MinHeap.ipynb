# Tests comparing two Minimum Heap data structures

```python
from MinHeapArray import exec_time_construction_array, exec_time_ajout_iteratif_array, exec_time_union_array, exec_time_insert_array,exec_time_extract_min_array
from MinHeapTree import exec_time_construction_tree, exec_time_ajout_iteratif_tree, exec_time_union_tree, exec_time_insert_tree,exec_time_extract_min_tree
import matplotlib.pyplot as plt
```

## Construction, Ajout Iteratif, Union

```python
avg_times_construction_array = exec_time_construction_array()
avg_times_insertKeyList_array = exec_time_ajout_iteratif_array()
avg_times_union_array = exec_time_union_array()

avg_times_construction_tree = exec_time_construction_tree()
avg_times_insertKeyList_tree = exec_time_ajout_iteratif_tree()
avg_times_union_tree = exec_time_union_tree()
```

## Result

```python
# BY FUNCTION
nbJeu = 5
nbCle = [1000, 5000, 10000, 20000, 50000, 80000, 120000, 200000]

fig = plt.figure(figsize=(15, 5))

# CONSTRUCTION
plt.subplot(1, 3, 1)  # row 1, col 2 index 1
plt.plot(nbCle, avg_times_construction_tree, label="Tree",
         marker='x', linestyle='-', color='b')
plt.plot(nbCle, avg_times_construction_array, label="Array",
         marker='x', linestyle='-', color='r')
plt.xlabel('Size of Data (Number of Keys)')
plt.ylabel('Average Execution Time (seconds)')
plt.legend()
plt.title('Construction')

# AJOUT ITERATIF
plt.subplot(1, 3, 2)  # index 2
plt.plot(nbCle, avg_times_insertKeyList_tree, label="Tree",
         marker='x', linestyle='-', color='b')
plt.plot(nbCle, avg_times_insertKeyList_array,
         label="Array", marker='x', linestyle='-', color='r')
plt.xlabel('Size of Data (Number of Keys)')
plt.ylabel('Average Execution Time (seconds)')
plt.legend()
plt.title('AJOUT ITERATIF')

# UNION
plt.subplot(1, 3, 3)  # index 2
plt.plot(nbCle, avg_times_union_tree, label="Tree",
         marker='x', linestyle='-', color='b')
plt.plot(nbCle, avg_times_union_array, label="Array",
         marker='x', linestyle='-', color='r')
plt.xlabel('Size of Data (Number of Keys)')
plt.ylabel('Average Execution Time (seconds)')
plt.legend()
plt.title('UNION')
plt.show()
```
